Transformamos el GRD completo para poder hacer el segundo merge y obtener los tratamientos aplicados a cada paciente.

```python
import pandas as pd
import numpy as np
df = pd.read_parquet(r'/home/brunoc/Documents/Universidad/2025/Taller de obtención y preparación de datos/Proyecto/Datos/Datos/GRD.parquet')
```

```python
pd.set_option('display.max_columns', None)
```

```python
df.head()
```

```python
procedimiento = ['CIP_ENCRIPTADO']
for i in range(1, 31):    
    procedimiento.append('PROCEDIMIENTO' + str(i))
```

```python
df_procedimiento = df[procedimiento]
df_procedimiento['CIP_ENCRIPTADO'].info()
```

```python
def replace(row):
    if row == "DESCONOCIDO":
        return np.nan
    return row
    
df["CIP_ENCRIPTADO"] = df["CIP_ENCRIPTADO"].apply(replace)
df["CIP_ENCRIPTADO"] = df["CIP_ENCRIPTADO"].astype(float)
df.info()
```
